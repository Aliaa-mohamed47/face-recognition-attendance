## Import libraries

```python
import pickle
import numpy as np
import csv
import os
from datetime import datetime
from sklearn.preprocessing import normalize
```

## Load Model

```python
MODEL_PATH = "../outputs/models/svm_classifier.pkl"

with open(MODEL_PATH, "rb") as f:
    data_model = pickle.load(f)

model = data_model["model"]
label_encoder = data_model["label_encoder"]

print("Model Loaded")
```

## Predict Function

```python
def predict_face(embedding):
    embedding = np.array(embedding).reshape(1, -1)
    embedding = normalize(embedding, norm="l2")

    pred = model.predict(embedding)[0]
    name = label_encoder.inverse_transform([pred])[0]

    return name
```

## Attendance System

```python
def mark_attendance(name):
    today = datetime.now().strftime("%Y-%m-%d")
    file_name = f"../outputs/logs/attendance_{today}.csv"

    os.makedirs("../outputs", exist_ok=True)

    already_marked = False

    if os.path.isfile(file_name):
        with open(file_name, "r") as f:
            reader = csv.reader(f)
            for row in reader:
                if row and row[0] == name:
                    already_marked = True
                    break

    if not already_marked:
        with open(file_name, "a", newline="") as f:
            writer = csv.writer(f)

            if os.stat(file_name).st_size == 0:
                writer.writerow(["Name", " Time"])

            time_now = datetime.now().strftime("%I:%M:%S %p").lstrip("0")
            writer.writerow([name, " " + time_now])

        print(f"Attendance marked for {name}")
    else:
        print(f"{name} already marked")
```

## Main Pipeline

```python
def process_face(embedding):
    name = predict_face(embedding)
    print(f"Prediction: {name}")
    mark_attendance(name)
```

## Load Embeddings

```python
with open("../outputs/database/face_embeddings.pkl", "rb") as f:
    data = pickle.load(f)

print("Embeddings Loaded")
```

## Checking for attendance

```python
for person, embeddings in data.items():
    print(f"\nTesting: {person}")

    embedding = np.array(embeddings["mean"])

    process_face(embedding)
```

## Import libraries

```python
import pickle
import numpy as np
import csv
import os
from datetime import datetime
from sklearn.preprocessing import normalize
```

## Load Model

```python
BASE_DIR = r"D:\face-recognition-attendance"

MODEL_PATH = os.path.join(BASE_DIR, "outputs", "models", "svm_classifier.pkl")
LOG_DIR    = os.path.join(BASE_DIR, "outputs", "logs")
DB_PATH    = os.path.join(BASE_DIR, "outputs", "database", "face_embeddings.pkl")

os.makedirs(LOG_DIR, exist_ok=True)
def load_model(path):
    with open(path, "rb") as f:
        data = pickle.load(f)

    print("Model Loaded ✅")
    return data["model"], data["label_encoder"]
```

## Predict Function

```python
def predict_face(embedding, model, label_encoder):
    embedding = np.array(embedding).reshape(1, -1)
    embedding = normalize(embedding, norm="l2")

    pred = model.predict(embedding)[0]
    name = label_encoder.inverse_transform([pred])[0]

    return name
```

## Attendance System

```python
def mark_attendance(name, log_dir):
    today = datetime.now().strftime("%Y-%m-%d")
    file_path = os.path.join(log_dir, f"attendance_{today}.csv")

    # check duplicate
    if os.path.isfile(file_path):
        with open(file_path, "r") as f:
            for row in csv.reader(f):
                if row and row[0] == name:
                    print(f"{name} already marked")
                    return False

    write_header = not os.path.isfile(file_path) or os.stat(file_path).st_size == 0

    with open(file_path, "a", newline="") as f:
        writer = csv.writer(f)

        if write_header:
            writer.writerow(["Name", "Time", "Date"])

        time_now = datetime.now().strftime("%I:%M:%S %p")
        writer.writerow([name, time_now, today])

    print(f"Attendance marked for {name}")
    return True
```

## Main Pipeline

```python
def process_face(embedding, model, label_encoder, log_dir):
    name = predict_face(embedding, model, label_encoder)
    print(f"Prediction: {name}")
    mark_attendance(name, log_dir)
    return name
```

## Load Embeddings

```python
with open(DB_PATH, "rb") as f:
    data = pickle.load(f)

print("Embeddings Loaded")
```

## Checking for attendance

```python
model, label_encoder = load_model(MODEL_PATH)

for person, embeddings in data.items():
    print(f"\nTesting: {person}")

    embedding = np.array(embeddings["mean"])

    process_face(embedding, model, label_encoder, LOG_DIR)
```
